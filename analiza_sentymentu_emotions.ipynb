## Zrozumienie biznesowe

Firma zajmująca się **monitoringiem mediów społecznościowych** stoi przed wyzwaniem **automatyzacji procesu analizy emocji** w treściach publikowanych przez użytkowników. Dotychczasowa manualna kategoryzacja tysięcy postów dziennie jest czasochłonna i kosztowna, ograniczając możliwość reakcji w czasie rzeczywistym na zmiany nastrojów konsumentów.

Celem projektu jest stworzenie modelu **automatycznej klasyfikacji sześciu podstawowych emocji**: radości, smutku, złości, strachu, miłości i zaskoczenia. Model powinien osiągnąć **minimum 75% dokładności**, umożliwiając praktyczne zastosowanie jako narzędzie wspomagające analityków.

Implementacja systemu przyniesie wartość poprzez szybką **identyfikację potencjalnych kryzysów wizerunkowych**, **monitoring skuteczności kampanii marketingowych** oraz **optymalizację alokacji zasobów analitycznych**. To pozwoli firmie oferować klientom bardziej responsywne i dokładne usługi monitoringu nastrojów w mediach społecznościowych.

## Przetwarzanie danych

```python
(X_train, X_val, X_test, y_train, y_val, y_test), preprocessor, stats = prepare_data()
```

Po zastosowaniu **TF-IDF wektoryzacji** z inteligentnym filtrowaniem stop words otrzymaliśmy macierze cech gotowe do modelowania. Automatycznie wykryto **1619 słów niecharakterystycznych** występujących w >66% klas emocji, co pozwoliło zachować słowa rzeczywiście różnicujące emocje. TF-IDF (Term Frequency-Inverse Document Frequency) nadaje wysokie wagi słowom **częstym w konkretnym tekście ale rzadkim w całym zbiorze danych**, wydobywając charakterystyczne cechy każdej emocji.

Przetwarzanie spowodowało utratę **~10% próbek** (głównie bardzo krótkie teksty), pozostawiając **14,492 teksty treningowe**. Utworzono słownik 5000 najważniejszych uni- i bigramów, gdzie każdy tekst jest reprezentowany jako wektor TF-IDF scores.

Top cechy wg średniego TF-IDF score to **charakterystyczne słowa emocji**: "agitated" (0.0052), "hated" (0.0049), "generous" (0.0039), "depressed" (0.0039), potwierdzając skuteczność filtrowania. Bigramy stanowią mniejszość ale dodają kontekst.

Dane są gotowe do trenowania modeli klasyfikacji z macierzami rzadkimi o wymiarach (próbki × 5000 cech).

## Modelowanie

```python
models, best_model, comparison = train_and_compare_models(X_train, y_train, X_val, y_val)
```

Porównanie trzech algorytmów ujawniło wyraźne różnice w wydajności i overfittingu. **Logistic Regression** okazał się najlepszy z walidacyjną dokładnością **80.93%** i deltą overfittingu **9.77%**, przewyższając **Naive Bayes** (78.72%, delta 7.39%) i **Random Forest** (78.00%, delta 16.94%).

Analiza F1 per emocja pokazuje mieszane wyniki. ## Wnioski i rekomendacje

Model osiągnął **82% dokładności w klasyfikacji emocji**, co przewyższa założony próg 75%. Szczególnie skuteczne okazało się rozpoznawanie **sadness (F1: 0.873)** i **anger (F1: 0.846)** - kluczowych emocji dla monitoringu kryzysów wizerunkowych. Wysoka czułość **joy (recall: 0.932)** pozwala na efektywne **wychwytywanie pozytywnych reakcji** na kampanie marketingowe.

Analiza confusion matrix ujawnia intrygujące wzorce błędów. Joy jest najczęściej mylone z sadness (76 przypadków), co zaskakuje biorąc pod uwagę przeciwną naturę tych emocji. To nie wynika z popularności sadness, ale prawdopodobnie **z obecności ironii/sarkazmu** ("I feel *great* about failing"), **mieszanych emocji** ("happy but sad to leave") oraz **kontekstów nostalgicznych**. Surprise wykazuje najgorsze wyniki (F1: 0.673) z rozprzestrzeniem błędów po wszystkich klasach, co wskazuje na **fundamentalny problem z definicją tej emocji** w danych.

Analiza top cech ujawnia, że użytkownicy mediów społecznościowych często używają **wulgaryzmów do wyrażania silnych emocji** ("fucked" w anger, "shitty" w sadness), co może powodować problemy w profesjonalnych kontekstach biznesowych. Dodatkowo, **wiele wyrażeń jest dwuznacznych** - "shaky" może oznaczać strach, ale też fizyczne drżenie z zimna, "superior" może wyrażać radość z sukcesu lub arogancję. Te subtelności językowe stanowią wyzwanie dla automatycznej klasyfikacji i wymagają świadomości kontekstu przy interpretacji wyników.

Kluczowe jest lepsze radzenie sobie z kontekstem i wieloznacznością. Dobrym pomysłem wydaje się rozważenie **pre-processingu wykrywającego ironię**, dodanie kategorii **"mixed emotions"**, oraz **targeted sampling** dla surprise. **Fine-tuning progu decyzyjnego** dla joy może zwiększyć precyzję kosztem recall. Model ujawnił złożoność wyrażania emocji w mediach społecznościowych, gdzie pozornie przeciwne uczucia często współwystępują, a język potoczny i wulgaryzmy są naturalnym sposobem wyrażania intensywnych emocji.
